```python
preguntas = [
    {
        "pregunta": "¿Qué artista lanzó el éxito “...Baby One More Time” en 1998?",
        "opciones": ["A) Christina Aguilera", "B) Britney Spears", "C) Jessica Simpson", "D) Avril Lavigne"],
        "respuesta": "B"
    },
    {
        "pregunta": "¿Quién es conocido como el 'Rey del Pop'?",
        "opciones": ["A) Michael Jackson", "B) Elvis Presley", "C) Prince", "D) Justin Timberlake"],
        "respuesta": "A"
    },
    {
        "pregunta": "¿Qué grupo británico lanzó el álbum 'Abbey Road'?",
        "opciones": ["A) Queen", "B) The Beatles", "C) The Rolling Stones", "D) Oasis"],
        "respuesta": "B"
    },
    {
        "pregunta": "¿Cuál es el nombre real de Lady Gaga?",
        "opciones": ["A) Stefani Germanotta", "B) Robyn Fenty", "C) Katheryn Hudson", "D) Alecia Moore"],
        "respuesta": "A"
    },
    {
        "pregunta": "¿Qué cantante popularizó la canción 'Shape of You'?",
        "opciones": ["A) Ed Sheeran", "B) Sam Smith", "C) Shawn Mendes", "D) Bruno Mars"],
        "respuesta": "A"
    },
    {
        "pregunta": "¿Quién interpretó el tema principal de la película 'Titanic'?",
        "opciones": ["A) Mariah Carey", "B) Celine Dion", "C) Whitney Houston", "D) Shania Twain"],
        "respuesta": "B"
    },
    {
        "pregunta": "¿Qué banda lanzó el álbum 'A Night at the Opera'?",
        "opciones": ["A) Queen", "B) Pink Floyd", "C) Led Zeppelin", "D) U2"],
        "respuesta": "A"
    },
    {
        "pregunta": "¿Cuál es el nombre artístico de Abel Tesfaye?",
        "opciones": ["A) Drake", "B) The Weeknd", "C) Post Malone", "D) Usher"],
        "respuesta": "B"
    },
    {
        "pregunta": "¿Qué artista lanzó el álbum '1989'?",
        "opciones": ["A) Ariana Grande", "B) Taylor Swift", "C) Selena Gomez", "D) Demi Lovato"],
        "respuesta": "B"
    },
    {
        "pregunta": "¿Quién canta 'Rolling in the Deep'?",
        "opciones": ["A) Adele", "B) Beyoncé", "C) Rihanna", "D) Alicia Keys"],
        "respuesta": "A"
    }
]

puntaje = 0

for i, q in enumerate(preguntas, 1):
    print(f"\nPregunta {i}: {q['pregunta']}")
    for opcion in q["opciones"]:
        print(opcion)
    respuesta = input("Tu respuesta: ").upper()
    if respuesta == q["respuesta"]:
        print("¡Correcto!")
        puntaje += 1
    else:
        print("Incorrecto.")

print(f"\nTu puntaje final es: {puntaje} de {len(preguntas)}")
```

```python
def hacer_pregunta(numero):
    global puntaje  # Accede a la variable global 'puntaje'
    pregunta = preguntas[numero]
    print(f"\nPregunta {numero + 1}: {pregunta['pregunta']}")
    for opcion in pregunta["opciones"]:
        print(opcion)
    respuesta = input("Tu respuesta (A, B, C o D): ").strip().upper()
    if respuesta == pregunta["respuesta"]:
        print("¡Correcto!")
        puntaje += 1
    else:
        print(f"Incorrecto. La respuesta correcta era {pregunta['respuesta']}.")

i = 0
while i < len(preguntas):
    hacer_pregunta(i)
    i += 1

print(f"\nHas terminado la trivia. Tu puntaje final es: {puntaje} de {len(preguntas)}.")
input("\nPresiona Enter para salir...")
```

```python
import random

opciones = ["piedra", "papel", "tijera", "lagarto", "spock"]
print("¡Bienvenido a Piedra, Papel, Tijera, Lagarto o Spock!")
usuario = input("Elige: piedra, papel, tijera, lagarto o spock ➡️ ").lower()
computadora = random.choice(opciones)
print(f"Tú elegiste: {usuario}")
print(f"La computadora eligió: {computadora}")

gana = {
    "tijera": ["papel", "lagarto"],
    "papel": ["piedra", "spock"],
    "piedra": ["lagarto", "tijera"],
    "lagarto": ["spock", "papel"],
    "spock": ["tijera", "piedra"]
}

if usuario == computadora:
    print("¡Empate!")
elif computadora in gana.get(usuario, []):
    print("¡Ganaste! 🎉")
else:
    print("Perdiste 😢")
```
