```python
def hacer_pregunta(numero):
    global puntaje  # Accede a la variable global 'puntaje'
    pregunta = preguntas[numero]
    print(f"\nPregunta {numero + 1}: {pregunta['pregunta']}")
    for opcion in pregunta["opciones"]:
        print(opcion)
    respuesta = input("Tu respuesta (A, B, C o D): ").strip().upper()
    if respuesta == pregunta["respuesta"]:
        print("Â¡Correcto!")
        puntaje += 1
    else:
        print(f"Incorrecto. La respuesta correcta era {pregunta['respuesta']}.")

i = 0
while i < len(preguntas):
    hacer_pregunta(i)
    i += 1

print(f"\nHas terminado la trivia. Tu puntaje final es: {puntaje} de {len(preguntas)}.")
input("\nPresiona Enter para salir...")
```
